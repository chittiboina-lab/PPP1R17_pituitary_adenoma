```python
# Western + PCR from mCorts, PPP1R17
# proliferation, cell cycle
# mCort vs GFP RNAseq
# mCort total/phospho proteome
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## mCort_GFP vs mCort_PPP1R17 RNAseq heatmap

```python
data = pd.read_csv('../../mCort_bulkRNAseq/selected.csv',index_col=0)
# data = data.dropna()
names = ['mCort_GFP rep1','mCort_GFP rep2','mCort_GFP rep3',
         'mCort_R17 rep1','mCort_R17 rep2','mCort_R17 rep3']
data.columns = names
data = data.transpose()
data
```

```python
fig = sns.clustermap(data, 
               z_score = 1,
               row_cluster=True, col_cluster=True,
#                cmap=greenpurp,
#                cmap=bluepink,
               cmap='icefire',
               figsize=(16,7),
               dendrogram_ratio=0.1,
               cbar_pos=(1.06,0.65,0.05,0.25)
              )
fig.ax_col_dendrogram.set_visible(False)
fig.ax_heatmap.set_yticks(ticks = fig.ax_heatmap.get_yticks(), labels = fig.ax_heatmap.get_yticklabels(), fontsize=16, fontweight = "bold")
fig.ax_heatmap.set_xticks(ticks = fig.ax_heatmap.get_xticks(), labels = '')
fig.ax_heatmap.tick_params(bottom = False)
fig.fig.suptitle('mCort_GFP vs mCort_R17: Bulk RNAseq', fontsize=24, fontweight='bold')
fig.savefig('mcort_bulkseq.png')
```

## mCort_GFP vs mCort_PPP1R17 mass spec data (total and phospho proteome)

```python
# total Proteins, horizontal
data = pd.read_csv('../../proteome_analysis/vis_data/mp_selected_total.txt', sep = '\t',index_col=0)
names = ['mCort_GFP rep1','mCort_GFP rep2','mCort_GFP rep3','mCort_GFP rep4','mCort_GFP rep5',
         'mCort_R17 rep1','mCort_R17 rep2','mCort_R17 rep3','mCort_R17 rep4','mCort_R17 rep5']
data.columns = names
data = data.transpose()
data
```

```python
fig = sns.clustermap(data, 
               z_score = 1,
               row_cluster=True, col_cluster=True,
#                cmap=greenpurp,
#                cmap=bluepink,
               cmap='rocket',
               figsize=(16,6),
               dendrogram_ratio=0.1,
               cbar_pos=(1.06,0.65,0.05,0.25)
              )
fig.ax_col_dendrogram.set_visible(False)
fig.ax_heatmap.set_yticks(ticks = fig.ax_heatmap.get_yticks(), labels = fig.ax_heatmap.get_yticklabels(), fontsize=16, fontweight = "bold")
fig.ax_heatmap.set_xticks(ticks = fig.ax_heatmap.get_xticks(), labels = '')
fig.ax_heatmap.tick_params(bottom = False)
fig.fig.suptitle('mCort_GFP vs mCort_R17: Total Proteins', fontsize=24, fontweight='bold')
fig.savefig('mcort_total_proteins.png')
```
