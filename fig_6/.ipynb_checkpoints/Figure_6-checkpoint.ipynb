```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## mCort GFP vs R17 vs R17+Fingolimod phospho proteome heatmap

```python
# Phospho Proteins
data = pd.read_csv('../../proteome_analysis/vis_data/f_selected_phospho.txt', sep = '\t',index_col=0)
data = data.drop(['GFP1', 'GFP2', 'GFP3', 'GFP4', 'GFP5'], axis=1)
names = ['mCort_R17 rep1','mCort_R17 rep2','mCort_R17 rep3','mCort_R17 rep4','mCort_R17 rep5',
         'mCort_R17+Fingolimod rep1','mCort_R17+Fingolimod rep2','mCort_R17+Fingolimod rep3','mCort_R17+Fingolimod rep4','mCort_R17+Fingolimod rep5',]
data.columns = names
data = data.transpose()
data
```

```python
# cmap = sns.diverging_palette(250, 30, l=65, center="dark", as_cmap=True)
# cmap = sns.diverging_palette(220, 20, l=60, s=90, center="dark", as_cmap=True)
fig = sns.clustermap(data, 
               z_score = 1,
               row_cluster=True, col_cluster=True,
               cmap='mako',
#                cmap=cmap,
#                metric="euclidean",
               figsize=(16,6),
               dendrogram_ratio=0.1,
               cbar_pos=(1.02,0.2,0.05,0.25)
              )
fig.ax_col_dendrogram.set_visible(False)
fig.ax_heatmap.set_yticks(ticks = fig.ax_heatmap.get_yticks(), labels = fig.ax_heatmap.get_yticklabels(), fontweight = "bold",  fontsize=16)
fig.ax_heatmap.set_xticks(ticks = fig.ax_heatmap.get_xticks(), labels = '')
fig.ax_heatmap.tick_params(bottom = False)
fig.fig.suptitle('mCort R17 vs mCort R17 + Fingolimod: Phospho Proteins', fontsize=24, fontweight='bold')
fig.savefig('fing_phospho.png')
```
