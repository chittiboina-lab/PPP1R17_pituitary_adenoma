## volcano plots

```python
peaks = pd.read_csv('../../results/ATAC/differential/DAC_peaks.csv', index_col=0)
x = peaks['log2FoldChange']
y = -np.log10(peaks['pvalue'])

up = {'x':[],'y':[]}
down = {'x':[],'y':[]}
non_sig = {'x':[],'y':[]}

blue = '#00A5CF'
red = '#D90368'
grey = '#d6d6d6'

for n in range(0,len(x)):
    if x[n] > 0.5 and y[n] > -log(0.05, 10):
        up['x'].append(x[n])
        up['y'].append(y[n])
    elif x[n] < -0.5 and y[n] > -log(0.05, 10):
        down['x'].append(x[n])
        down['y'].append(y[n])
    else:
        non_sig['x'].append(x[n])
        non_sig['y'].append(y[n])
```

```python
#Initialize Canvas
fig, ax = plt.subplots(1,1,figsize=(6,6))
#Plot Data
ax.scatter(non_sig['x'],non_sig['y'],
           c = grey,
           alpha=0.25,
          )
ax.scatter(up['x'],up['y'],
           c = red,
           alpha=0.25,
          )
ax.scatter(down['x'],down['y'],
           c = blue,
           alpha=0.25,
          )
#Cutoff Lines
ax.hlines(-log(0.05, 10),-6,6, linestyles="dotted", colors='black')
ax.vlines(-0.5,0,3.5, linestyles="dotted", colors='black')
# ax.vlines(0,0,3.5, linestyles="dotted", colors='black')
ax.vlines(0.5,0,3.5, linestyles="dotted",  colors='black')
#Titles
fig.suptitle('Differentially Accessible Peaks',x=0.52, fontsize=16, fontweight='bold')
ax.set_title('Tumor vs. Normal', fontsize=16)
ax.set_xlabel("Log2 Fold Change", fontsize=15)
ax.set_ylabel("-log10 P Value", fontsize=15)

# Plot Legend
ax.plot(3.5,3.4, marker="o", markersize=10,color=red, alpha=0.75,mec='black')
ax.text(3.85,3.36,'Up: %d' % (len(up['x'])), size=12, weight='bold')
ax.plot(3.5,3.2, marker="o", markersize=10,color=blue, alpha=0.75,mec='black')
ax.text(3.85,3.16,'Down: %d' % (len(down['x'])), size=12, weight='bold')

plt.savefig('DAC_volcano_peaks.png', dpi=600, bbox_inches='tight')
```

## ATAC Coverage Plot

```python
def tabix_query(filename, chrom, start, end, barcodes):
    query = '{}:{}-{}'.format(chrom, start, end)
    process = Popen(['tabix', filename, query], stdout=PIPE, universal_newlines=True)
    result = []
    for line in process.stdout:
        line = line.split()
        if len(barcodes) > 0:
            if line[3] in barcodes:
                result.append(line)
        elif len(barcodes) == 0:
            result.append(line)
    return result
#     result = []
#     process = Popen(['tabix', filename, query], stdout=PIPE, universal_newlines=True)
#     process = process.communicate()
#     if process[0] == '':
#         result.append(0)
#     elif process[0] != '':
#         frags_df = pd.read_csv(io.StringIO(process[0]), sep = '\t',header=None).drop([4], axis=1)
#         if len(barcodes) > 0:
#             frags_df = frags_df.loc[(frags_df[3].isin(barcodes))]
#         result.append()
#     return result

def coverage_plot(ax,
                  frag_file_s,        # string or list of strings specifying a path to a bed style fragments file to plot from: https://support.10xgenomics.com/single-cell-atac/software/pipelines/latest/output/fragments
                  gene,
                  extend_by=0,
                  barcodes=[],        # FOR NOW: a list of barcodes of cells of interest. Future plan to integrate with scanpy/anndata object
                  ylim = None,        # Y axis upper limit. Defualts to 110% of the max coverage value
                  c='royalblue',      # any color accepted by matplotlib
                  x_label=None,       # If not specified label will be 'Position', can be an empty string for no label
                  y_label=None,       # If not specified label will be 'Coverage', can be an empty string for no label
#                   y_ticks=
                  title=None,
                  scale_factor=1,
                  x_tick_labels = True,
                  y_tick_labels = False,
                 ):
    
    
    gene = ensembldb.genes_by_name(gene)[0]
    start = gene.start-extend_by
    end = gene.end+extend_by
    
    if len(barcodes) > 1:
        scale_factor=len(barcodes)
#     print(scale_factor, type(scale_factor))
    
    if type(frag_file_s) == list:
        frags = []
        for file in frag_file_s:
            frags += tabix_query(file, 'chr%s'%(gene.contig),str(start),str(end),barcodes)
    elif type(frag_file_s) == str:
        frags = tabix_query(file, 'chr%s'%(gene.contig),str(start),str(end),barcodes)
    
    x = np.arange(start,end+1)
    y = np.zeros(len(x))
    
    for frag in frags:
        y[int(frag[1])-start:int(frag[2])-start+1] += 1
    
    y = y/scale_factor
    
    out = ax.fill_between(x, y, color = c)
    
    if ylim == None:
        ax.set_ylim(0,max(y)+(max(y)/10))
    else:
        ax.set_ylim(0,ylim)
    ax.set_xlim(start,end)
    ax.set_frame_on(False)
    
    if y_label == None:
        ax.set_ylabel("Coverage", fontsize=12)
    elif type(y_label)==str:
        ax.set_ylabel("%s"%(y_label), fontsize=12)
    if y_tick_labels == False:
        ax.set_yticks([])
    if x_label == None:
        ax.set_xlabel("Position")
        ax.set_xticks([gene.start,gene.end])
        ax.set_xticklabels(['Start','End'],fontdict={'fontweight':'bold','size':10})
    elif type(x_label)==str:
        ax.set_xlabel("%s"%(x_label))
        ax.set_xticks([gene.start,gene.end])
    if x_tick_labels == True:
        ax.set_xticklabels(['Start','End'],fontdict={'fontweight':'bold','size':10})
    elif x_tick_labels == False:
        ax.set_xticks([])
    
    if title == None:
        ax.set_title("%s" % (gene.gene_name), fontweight="bold", fontsize=18)
    elif type(title)==str:
        ax.set_title("%s" % (title), fontweight="bold", fontsize=18)
    
    return out
    
```
