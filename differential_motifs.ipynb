```python
import pandas as pd
```

```python
up = pd.read_csv('up_motifs.txt', sep='\t', index_col=0)
up
```

```python
down = pd.read_csv('down_motifs.txt', sep='\t', index_col=0)
down
```

```python
up_index = set(up.index)
down_index = set(down.index)
diff_up = up_index.difference(down_index)
diff_up = list(diff_up)
diff_up
```

```python
diff_up = up.loc[diff_up]
diff_up
```

```python
diff_up.to_csv('differentially_accessible_motifs.txt', sep='\t')
```
