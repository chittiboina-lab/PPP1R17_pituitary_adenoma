```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../../methylation_analysis/results/beta_sig_01.csv', sep = ',',index_col=0)
names = ["CD 1", "CD 2", "CD 3", "CD 4", "CD 5", "CD 6", "Normal 1", "Normal 2", "Normal 3", "Normal 4", "Normal 5", "Normal 6", "Normal 7", "Normal 8", "Normal 9", "Normal 10", "Normal 11"]
data.columns = names
# data.dropna()
data
```

```python
data = data.dropna()
data = data.transpose()
data
```

```python
fig = sns.clustermap(data, 
               z_score = 1,
               row_cluster=True, col_cluster=True,
               cmap='icefire',
               figsize=(14,6),
               dendrogram_ratio=0.1,
               cbar_pos=(1.06,0.65,0.05,0.25)
              )
fig.ax_col_dendrogram.set_visible(False)
fig.ax_heatmap.set_yticks(ticks = fig.ax_heatmap.get_yticks(), labels = fig.ax_heatmap.get_yticklabels(), fontsize=16, fontweight = "bold")
fig.ax_heatmap.set_xticks(ticks = fig.ax_heatmap.get_xticks(), labels = '')
fig.ax_heatmap.tick_params(bottom = False)
fig.fig.suptitle("Differentially Methylated Regions: CD Adenoma vs Autopsy Derived Normal Gland", fontsize=20, fontweight='bold')
fig.savefig('methylation_heatmap.pdf', bbox_inches='tight')
```
